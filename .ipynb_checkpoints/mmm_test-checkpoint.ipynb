```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from matplotlib.ticker import PercentFormatter

def ROAS_curve( m, b, x ):
    roas = m*x+b
    if roas > 0:
        return roas / 100
    return None

x_range = [ x for x in range( 1, 365) ]

roas_curve_1 = [ ROAS_curve( -2, 300, x ) for x in x_range ]
roas_curve_2 = [ ROAS_curve( -1.25, 200, x ) for x in x_range ]

plt.plot(x_range, roas_curve_1, color='mediumseagreen', label='Channel 1' )
plt.plot(x_range, roas_curve_2, color='navy', label='Channel 2' )
plt.axhline( y=0, color='black', linestyle='dashed' )
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.gca().xaxis.set_major_formatter('${x:1.0f}')

plt.title("ROAS vs. Ad Spend", loc = 'center')
plt.xlabel("Ad Spend ($000s)")
plt.ylabel("ROAS (%)")

plt.legend(loc="upper right")

max_profit = 0
max_profit_curve_1 = 0
max_profit_curve_2 = 0

#profit max
for i in x_range:
    for z in x_range:
        if roas_curve_1[ i-1 ] is not None:
            profit_curve_1 = roas_curve_1[ i-1 ] * i - i
        if roas_curve_2[ z-1 ] is not None:
            profit_curve_2 = roas_curve_2[ z-1 ] * z - z
        if profit_curve_1 + profit_curve_2 > max_profit:
            max_profit = profit_curve_1 + profit_curve_2
            max_profit_curve_1 = i
            max_profit_curve_2 = z

plt.plot( max_profit_curve_1, roas_curve_1[ max_profit_curve_1 - 1], 'o', color='mediumseagreen' )
plt.plot( max_profit_curve_2, roas_curve_2[ max_profit_curve_2 - 1], 'o', color='navy' )
            
print( "Maximum Profit: " + " " + str( max_profit ) )
print( "Spend Level, Channel 1: " + " " + str( max_profit_curve_1 ) )
print( "Spend Level, Channel 2: " + " " + str( max_profit_curve_2 ) )
```
